```python
# - Importando as Bibliotecas de Machine Learning e Numpy e Matplotlib.
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
```

```python
# - Definindo o conjunto de Dados ( Manipular )
x1 = np.array([3, 1, 1, 2, 1, 6, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8])
x2 = np.array([5, 4, 6, 6, 5, 8, 6, 7, 6, 7, 1, 2, 1, 2, 3, 2, 3])
```

```python
# Exibindo as informações em formato de gráfico para análise.
plt.plot()
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Dataset - Análise com o K-Means')
plt.scatter(x1, x2)
plt.show()
```

```python
# create new plot and data
plt.plot()
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
colors = ['b', 'g', 'r']
markers = ['o', 'v', 's']
# KMeans algorithm 
K = 3
kmeans_model = KMeans(n_clusters=K).fit(X)

plt.plot()
for i, l in enumerate(kmeans_model.labels_):
    plt.plot(x1[i], x2[i], color=colors[l], marker=markers[l],ls='None')
    plt.xlim([0, 10])
    plt.ylim([0, 10])

plt.show()

```

```python
#Dataset IrisPython

import pandas as pd 
iris = pd.read_csv("iris(1).csv")
iris.head() #Mostae as 5 primeira linhas do DataFrame
```

```python
 # Remover a coluna de Espécie no final do data frame.
 X = iris.iloc[:, 0:4].values
X
```

```python
# Instanciar o objeto do KMeans 
kmeans = KMeans(n_clusters = 3, init = 'random')
```

```python
# Treinando Modelo de Machine Learning (KMeans)
kmeans.fit(X)
```

```python
# Exibir a quantidade de centroides de identificados.
kmeans.cluster_centers_
```

```python
# Exibir a distancia entre os pontos centrais
distance = kmeans.fit_transform(X)
distance
```

```python
# Exibir os rótulos do Cluster completo
labels = kmeans.labels_
labels
```

```python
# Método Elbow -> O método Elbow se trata de uma técnica interessante para encontrar o valor 
#ideal do parâmetro k.
# Basicamente o que o método faz é testar a variância dos dados em relação ao número de clusters.
```

```python
from sklearn.cluster import KMeans
wcss = []
i = 0 
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random')
    kmeans.fit(X)
    print (i,kmeans.inertia_)
    wcss.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss)
plt.title('O Metodo Elbow')
plt.xlabel('Numero de Clusters')
plt.ylabel('WSS') #within cluster sum of squares
plt.show()
```

```python
# Agrupando Novos Dados com KMeans
# Com o nosso modelo já criado podemos usar o algoritmo K-means para agrupar novos 
#dados baseado nos centroides que já estão definidos.
# Passando novos valores e observando os cluster que serão atribuídos.
```

```python
#Agrupando novos dados KMeansPython
data = [
        [ 4.12, 3.4, 1.6, 0.7],
        [ 5.2, 5.8, 5.2, 6.7],
        [ 3.1, 3.5, 3.3, 3.0]
    ]
kmeans.predict(data)
```

```python
# Visualizando os resultados
# Para ficar ainda mais claro é interessante visualizar através de uma representação
# gráfica os nossos dados e os centroides criados pelo K-means Veja com o código abaixo como é simples.
```

```python
import matplotlib.pyplot as plt
 
plt.scatter(X[:, 0], X[:,1], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'red',label = 'Centroids')
plt.title('Iris Clusters and Centroids')
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.legend()
 
plt.show()
```
